### CSV File

````python
#1. CSV File by built in csv module
import csv

# try:.
def file_reader_csv(file_path):
    with open(file_path, mode='r', newline='') as file:
        csv_reader = csv.reader(file)
        
        # Optionally, skip headers if needed
        # headers = next(csv_reader, None)
        
        for row in csv_reader:
            print(row)

# except FileNotFoundError:
#     print("Error: The specified CSV file was not found.")
# except PermissionError:
#     print("Error: Permission denied. Please check your file permissions.")
# except Exception as e:
#     print(f"An unexpected error occurred while reading the CSV file: {e}")
````

````python
file_reader_csv("GDP_Data/GDP_data_csv.csv")
````

```

Import csv: We import the csv module to handle CSV functionality.

with open('data.csv', mode='r', newline='') as file:: We open the file in read mode. 
The newline='' is often recommended in Python 3 when dealing with CSVs to prevent extra blank lines on some platforms.

csv.reader(file): This returns an iterable that we can loop over to get each row.

FileNotFoundError: Catches the scenario where the CSV file does not exist or the path is incorrect.

PermissionError: Occurs if you dont have the required permissions to read the file.

Exception as e: A catch-all for other unexpected errors.

5. Combining & Merging Data

````python
# Part 1: Concatenating DataFrames


# Sample DataFrame 1
df1 = pd.DataFrame({
    'Product': ['Apple', 'Banana', 'Cherry'],
    'Sales_Q1': [100, 150, 200]
})

# Sample DataFrame 2
df2 = pd.DataFrame({
    'Product': ['Date', 'Elderberry', 'Fig'],
    'Sales_Q1': [80, 120, 60]
})

print (df1)
print (df2)



````

````python
# Concatenate df1 and df2 by rows
df_concat = pd.concat([df1, df2], axis=0)
print("After Concatenation (Rows):\n", df_concat)

````

````python
# We can also concatenate by columns if we have common row indices.
# For demonstration, let's create new data with the same row indices:
df3 = pd.DataFrame({
    'Sales_Q2': [90, 110, 170]
}, index=[0, 1, 2])              # pay attention to key

print (df3)
````

````python
df_concat_cols = pd.concat([df1, df3], axis=1)
print("\nAfter Concatenation (Columns):\n", df_concat_cols)
````

````python
# ----------------------------
# Part 2: Merging DataFrames
# ----------------------------

# Sample DataFrame 4
df4 = pd.DataFrame({
    'Product': ['Apple', 'Banana', 'Cherry', 'Date'],
    'Price': [1.0, 0.5, 2.0, 3.0]
})

df4
````

````python
# Merge df_concat with df4 on "Product"
# Default is an inner join
df_merged_inner = pd.merge(df_concat, df4, on='Product', how='inner')   #on column name not on keys
print("\nMerged with an INNER join:\n", df_merged_inner)

````

````python
# Let's see how an outer join would work
df_merged_outer = pd.merge(df_concat, df4, on='Product', how='outer')
print("\nMerged with an OUTER join:\n", df_merged_outer)
````

````python
# ----------------------------
# Part 3: Example with different key columns
# ----------------------------

# DataFrame 5 with a different column name
df5 = pd.DataFrame({
    'Item': ['Apple', 'Cherry', 'Grape'],
    'Supplier': ['Company A', 'Company B', 'Company C']
})

df5
````

````python
# Merge on different columns using 'left_on' and 'right_on'
df_merged_diff_keys = pd.merge(df_concat, df5,
                               left_on='Product', 
                               right_on='Item', 
                               how='left')                           #use 'right' and see
print("\nMerged on different key columns:\n", df_merged_diff_keys)
````

```

Concatenation:
By rows (axis=0): Stacks DataFrames vertically, assuming they have the same or compatible columns.
By columns (axis=1): Combines DataFrames side by side, matching rows by index.


Merging (Similar to SQL joins):
* pd.merge() needs a key column or columns. By default, it merges on columns with the same name in both DataFrames.
* how='inner' returns only the matching rows in both DataFrames.
* how='outer' returns all rows from both DataFrames, filling missing values with NaN.
* You can use left_on and right_on to merge when the key columns in each DataFrame have different names.

6. Basic Plotting

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)  # For reproducible results
dates = pd.date_range("2023-01-01", periods=10, freq="D")
sales = np.random.randint(50, 150, size=10)          # Random sales numbers
customers = np.random.randint(5, 25, size=10)        # Random number of customers
temperature = np.random.uniform(10, 30, size=10)     # Random temperature values

# Create a pandas DataFrame
df = pd.DataFrame({
    "Date": dates,
    "Sales": sales,
    "Customers": customers,
    "Temperature": temperature
})

# Set "Date" as the index to facilitate time-series plots
df.set_index("Date", inplace=True)

# Display the first few rows
print("DataFrame Head:\n", df.head())



````

````python
# -------------------------
# Basic Line Plot
# -------------------------
plt.figure(figsize=(8, 4))
plt.plot(df.index, df["Sales"], marker='o', label="Sales")
plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

````

````python
# -------------------------
# Basic Bar Plot
# -------------------------
plt.figure(figsize=(8, 4))
plt.bar(df.index, df["Customers"], color='skyblue', label="Customers")
plt.title("Number of Customers Over Time")
plt.xlabel("Date")
plt.ylabel("Customers")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
````

````python

# -------------------------
# Basic Histogram
# -------------------------
plt.figure(figsize=(6, 4))
plt.hist(df["Sales"], bins=5, color='coral', alpha=0.7, edgecolor='black')
plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

````

````python
# -------------------------
# Basic Scatter Plot
# -------------------------
plt.figure(figsize=(6, 4))
plt.scatter(df["Customers"], df["Sales"], c=df["Temperature"], cmap='viridis', s=100)
plt.title("Sales vs. Customers")
plt.xlabel("Customers")
plt.ylabel("Sales")
cbar = plt.colorbar()
cbar.set_label("Temperature (C)")
plt.tight_layout()
plt.show()
````
